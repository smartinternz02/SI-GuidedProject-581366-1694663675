```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
crashes = sns.load_dataset("car_crashes")
sns.set(style="whitegrid")
```

```python
crashes
```

```python
sns.pairplot(crashes)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(crashes["total"], bins=20, kde=True)
plt.title("Distribution of Total Accidents")
plt.xlabel("Total Accidents")
plt.ylabel("Frequency")
plt.show()
```

```python
# Inference: The histogram for the distribution of the total number of accidents shows a right-skewed distribution, with most states having a lower number of accidents, clustered around 5 to 10 accidents.
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="total", y="abbrev", data=crashes, orient="h")
plt.title("Total Accidents by State")
plt.xlabel("Total Accidents")
plt.ylabel("State Abbreviation")
plt.show()
```

```python
# Inference: In the barplot showing the total number of accidents by state, Texas has the highest number of accidents, followed by Mississippi and Alabama.
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x="alcohol", y="speeding", data=crashes)
plt.title("Alcohol vs. Speeding Accidents")
plt.xlabel("Alcohol-Related Accidents")
plt.ylabel("Speeding-Related Accidents")
plt.show()
```

```python
# Inference: The scatterplot comparing alcohol-related accidents to speeding-related accidents suggests a positive correlation between the two. States with higher alcohol-related accidents tend to have higher speeding-related accidents
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x="abbrev", data=crashes, palette="Set2")
plt.title("Count of Accidents by State")
plt.xlabel("State Abbreviation")
plt.ylabel("Count of Accidents")
plt.xticks(rotation=90)
plt.show()
```

```python
# Inference: The countplot of accidents by state reveals that Texas has the highest count of accidents, while other states vary in their accident counts.
```

```python
sns.jointplot(x="alcohol", y="speeding", data=crashes, kind="scatter")
plt.suptitle("Jointplot of Alcohol and Speeding Accidents")
plt.show()
```

```python
# Inference: The jointplot provides a visual representation of the relationship between alcohol and speeding accidents. It shows that as alcohol-related accidents increase, speeding-related accidents also tend to increase.
```

```python
sns.relplot(x="alcohol", y="speeding", data=crashes, kind="scatter")
plt.title("Relationship Between Alcohol and Speeding Accidents")
plt.xlabel("Alcohol-Related Accidents")
plt.ylabel("Speeding-Related Accidents")
plt.show()
```

```python
# Infernce: The relplot displaying the relationship between alcohol-related and speeding-related accidents visually confirms the positive correlation observed in the scatterplot.
```

```python
plt.figure(figsize=(8, 6))
sns.distplot(crashes["alcohol"], bins=10, kde=True)
plt.title("Distribution of Alcohol-Related Accidents")
plt.xlabel("Alcohol-Related Accidents")
plt.ylabel("Density")
plt.show()
```

```python
# Infernce: The distplot for alcohol-related accidents indicates a right-skewed distribution, suggesting that most states have a relatively lower number of alcohol-related accidents.
```

```python
correlation_matrix = crashes.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Between Accident Attributes")
plt.show()
```

```python
# Inference: The heatmap illustrating the correlation between accident attributes shows that there is a positive correlation between some attributes, such as "alcohol" and "total," indicating that states with more alcohol-related accidents tend to have higher total accidents.
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x="alcohol", data=crashes)
plt.title("Distribution of Alcohol-Related Accidents")
plt.xlabel("Alcohol-Related Accidents")
plt.show()
```

```python
# Inference: The boxplot for alcohol-related accidents indicates the presence of some outliers, suggesting that a few states have a significantly higher number of alcohol-related accidents compared to others.
```

```python
speeding_accidents = crashes["speeding"].sum()
non_speeding_accidents = crashes["total"].sum() - speeding_accidents
labels = ["Speeding", "Non-Speeding"]
sizes = [speeding_accidents, non_speeding_accidents]
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Percentage of Accidents Due to Speeding")
plt.axis("equal")
plt.show()
```

```python
# Inference: The pie chart displaying the percentage of accidents due to speeding shows that the majority of accidents are not primarily due to speeding, indicating that other factors contribute to a significant portion of accidents.
```
